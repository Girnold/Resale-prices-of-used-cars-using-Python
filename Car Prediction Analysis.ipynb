```python
headers = ["symboling", "normalized-losses", "make", 
           "fuel-type", "aspiration","num-of-doors",
           "body-style","drive-wheels", "engine-location",
           "wheel-base","length", "width","height", "curb-weight",
           "engine-type","num-of-cylinders", "engine-size", 
           "fuel-system","bore","stroke", "compression-ratio",
           "horsepower", "peak-rpm","city-mpg","highway-mpg","price"]

df.columns=headers
df.head()
```

```python
df.isna().any()

df.isnull().any()
```

```python
df['city-mpg'] = 235 / df['city-mpg']
df.rename(columns = {"city-mpg":"city-L/100km"},inplace=True)

print(df.columns)

df.dtypes
```

```python
df.price.unique()

df=df[df.price != '?']

df['price'] = df['price'].astype(int)

df.dtypes
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

```python
bins = np.linspace(min(df['price']),max(df['price']),4)
group_names = ['Low',"Medium",'High']
df['price-binned'] = pd.cut(df['price'],bins,
                           labels = group_names,
                           include_lowest = True)

print(df['price-binned'])
plt.hist(df['price-binned'])
plt.show()
```

```python
pd.get_dummies(df['fuel-type']).head()

df.describe()
```

```python
plt.boxplot(df['price'])

sns.boxplot(x = 'drive-wheels',y = 'price',data=df)

plt.scatter(df['engine-size'],df['price'])
plt.title('Scatterplot of Enginesize vs Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.grid()
plt.show()
```

```python
test = df[['drive-wheels','body-style','price']]
df_grp = test.groupby(['drive-wheels','body-style'],
                     as_index=False).mean()

df_grp
```

```python
data_pivot = df_grp.pivot(index = "drive-wheels",columns = 'body-style')

data_array = data_pivot.values

plt.pcolormesh(data_array,cmap="RdBu",shading='auto')
plt.colorbar()

plt.show()
```

```python
data_annova = df[['make','price']]
grouped_annova = data_annova.groupby(['make'])
annova_results_1 = sp.stats.f_oneway(grouped_annova.get_group('honda')['price'],
                                    grouped_annova.get_group('subaru')['price'])

print(annova_results_1)

sns.regplot(x = 'engine-size',y = 'price',data = df)
plt.ylin(0,)
```
